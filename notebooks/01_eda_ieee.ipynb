# IEEE-CIS EDA (Phase 1)

Real e-commerce transactions (Vesta). train_transaction left-joined with train_identity on
TransactionID. This notebook checks fraud rate, missingness, the time axis, categoricals, and
anomalies. Conclusions are written to `docs/eda.md`.

```python
from fraud_aml.data.loaders import load_merged

df = load_merged()
print(df.shape)
```

## Class balance (fraud rate)

```python
rate = df["isFraud"].mean()
print(f"fraud rate: {rate:.4f}  (~1:{(1 - rate) / rate:.0f})")
```

## Missingness (top-20 columns)

```python
missing = df.isna().mean().sort_values(ascending=False)
missing.head(20)
```

## Time axis (TransactionDT)

```python
dt = df["TransactionDT"]
print(f"span days: {(dt.max() - dt.min()) / 86400:.1f}")
df.assign(day=dt // 86400).groupby("day")["isFraud"].mean().plot(title="fraud rate by day")
```

## Categoricals

```python
for col in ["card4", "card6", "ProductCD", "P_emaildomain"]:
    print(col, "cardinality", df[col].nunique())
df.groupby("ProductCD")["isFraud"].mean()
```
